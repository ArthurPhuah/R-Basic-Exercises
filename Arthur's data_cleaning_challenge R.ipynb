```r
library(sjmisc)
library(dplyr)
library(tidyverse)
library(stringr)
```

```r
data_import = read.csv(file = "data_cleaning_challenge.csv")
data_import
```

```r
dim(data_import)
```

```r
# "subset" is used to drop variables
data_drop = subset(data_import, select = -c(X, X.1))
data_drop
```

```r
dim(data_drop)
```

```r
column_values = list()
iteration = 0

for (i in data_drop$Row.Type) {
    iteration = ifelse (str_contains(i, "first name:", ignore.case = TRUE), iteration + 1, iteration)      
    column_values = append(column_values, iteration)
}
data_drop$iteration = unlist(column_values)
data_drop
```

```r
dim(data_drop)
```

```r
left_data = data_drop[grepl('first name', data_drop$Row.Type),]
left_data = subset(left_data, select = -c(Speed1,Speed2,Electricity,Effort,Weight,Torque))
left_data
```

```r
dim(left_data)
```

```r
names(left_data)[names(left_data) == "Row.Type"] = "First Name"
names(left_data)[names(left_data) == "Iter.Number"] = "Last Name"
names(left_data)[names(left_data) == "Power1"] = "Date"
left_data$"First Name" = gsub("first name:", " ", as.character(left_data$"First Name"))
left_data$"Last Name" = gsub("last name:", " ", as.character(left_data$"Last Name"))
left_data$"Date" = gsub("date:", " ", as.character(left_data$"Date"))
left_data
```

```r
dim(left_data)
```

```r
right_data_drop = subset(data_drop, select = -c(Row.Type, Iter.Number))
right_data = right_data_drop[!right_data_drop$Speed1 == "", ]
right_data = right_data[!grepl('Speed1', right_data$Speed1), ]
right_data
```

```r
merge(left_data, right_data)
```

```r
dim(merge(left_data, right_data))
```
