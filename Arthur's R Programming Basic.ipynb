## 6. Write a R program to get all prime numbers up to a given number (based on the sieve of Eratosthenes)

```r
# Requirements:
# 1. Get all prime numbers up to a given number based on the sieve of Eratosthenes
# 2. Find out what is sieve of Eratosthenes
#
# "c()" is a generic function which combines its arguments to form a vector
# c(x, y) both x and y looks at the same number where x is the equation and y is print
# "function()" is a set of statements organized together to perform a specific task
# "if" is a statement to help with decision making in the program
# "==" is used to check whether 2 objects are equal
# "any()" is used to see a given logical vectors on whether at least one of the
# value is true
# "%%" is the modular division or integer-divide, it is used to divide 'x' by 'y'
# and return the remainder
```

```r
print("Sieve of Eratosthenes is an ancient algorithm for finding all prime number")
prime_numbers = function(n) {
    if(n >= 2) {
        x = seq(2, n)
        prime_nums = c()
        for (i in seq(2, n)) {
            if (any(x == i)) {
                prime_nums = c(prime_nums, i)
                x = c(x[(x %% i) != 0], i)
            }
        }
        return(prime_nums)
    }
    else {
        stop("Input number should be at least 2.")
    }
}
prime_numbers(100)
```

## 7. Write a R program to print the numbers from 1 to 100 and print "Fizz" for multiples of 3, print "Buzz" for multiples of 5, and print "FizzBuzz" for multiples of both

```r
# Requirements:
# 1. Print Fizz for multiples of 3
# 2. Print Buzz for multiples of 5
# 3. Print FizzBuzz for multiples of both
#
# You are printing numbers 1:100 and you have to replace the multiples of 3, 5 or 3 & 5
# in that list with Fizz, Buzz, or FizzBuzz
```

```r
for (i in seq(1,100)) {
    if (i %% 3 == 0 & i %% 5 == 0) {
        print("FizzBuzz")
        }
    else if (i %% 3 == 0) {
        print("Fizz")
    }
    else if (i %% 5 == 0) {
        print("Buzz")
    }
    else print(i)
}
```

## 9. Write a R program to find the factors of a given number

```r
# Requirements:
# 1. Find the factors of a given number
```

```r
factors = function(n) {
    print(paste("The factors of ", n, "are:"))
    
    for (i in 1:n) {
        if ((n %% i) == 0) {
            print(i)
        }
    }
}
factors(4)
factors(10)
factors(27)
```

## 21. Write a R program to create a simple bar plot of five subjects marks
